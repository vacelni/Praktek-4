# IIR vs FIR

## Sinyal dan Filter


Dalam bagian sebelumnya, kami memperkenalkan konsep sinyal, spektrum, dan bentuk Z. Selama proses perjalanan sinyal dari sumber ke penerima, urutan data mengalami perubahan dan transformasi. Transformasi ini disebabkan oleh perjalanan sinyal melalui berbagai perangkat komputasi yang menerapkan tugas matematika tertentu. Dalam tugas-tugas pemrosesan sinyal digital, data melewati rangkaian digital yang disebut filter. Filter digital, seperti filter analog, memiliki karakteristik yang berbeda. Biasanya, karakteristik ini diekspresikan dalam representasi frekuensi - yaitu karakteristik amplitudo-frekuensi dan fasa-frekuensi dari filter. Filter digital digunakan terutama untuk meningkatkan kualitas sinyal - untuk mengekstraksi sinyal dari urutan data, atau untuk memperburuk sinyal yang tidak diinginkan - untuk menekan sinyal tertentu dalam urutan sampel yang diterima.

Saati ini, filter digital digunakan hampir di mana-mana di mana pemrosesan sinyal diperlukan. Bahkan operasi matematika primitif (perkalian dan penambahan) dapat direpresentasikan sebagai filter digital. Oleh karena itu, semua perangkat yang mengelilingi kita (komputer, telepon, televisi, dll.) mengandung berbagai filter digital. Misalnya, televisi menggunakan filter untuk mengubah gambar yang ditampilkan. Penyetelan kecerahan - pada dasarnya adalah filtrasi gambar.
Dengan demikian, filter digital digunakan dalam berbagai tugas pemrosesan sinyal: analisis spektral, pemrosesan data audio dan video, pemrosesan ucapan, gerakan, dll. Secara khusus untuk tugas-tugas DSP, filter adalah sistem matematis tertentu yang mengubah bentuk sinyal masukan (amplitudo, fasa, frekuensi).

Keunggulan Filter Digital:

- Kemampuan untuk memperoleh karakteristik yang tidak mungkin diperoleh dengan metode analog (misalnya, respons frekuensi tepat linear).
- Filter digital lebih stabil terhadap perubahan lingkungan eksternal (suhu, kelembaban, tekanan). Oleh karena itu, filter digital tidak memerlukan kalibrasi yang terus-menerus.
- Kemudahan dalam mengubah frekuensi pemrosesan sinyal.
- Tergantung pada kecepatan pemrosesan dan frekuensi sinyal, satu filter dapat memproses data dari beberapa saluran secara bersamaan.
- Data sebelum dan setelah filter dapat disimpan dalam media penyimpanan apa pun. Sinyal analog sulit untuk direkam (dalam setiap kasus, konversi ke bentuk digital diperlukan dan sebaliknya).
- Konsumsi daya rendah dan biaya rendah relatif terhadap filter analog.
- Tingkat akurasi yang lebih tinggi dalam perhitungan (perangkat digital modern tidak terbatas oleh akurasi perhitungan).
- Kemudahan dalam merancang filter dengan karakteristik frekuensi yang kompleks.
- Dalam tugas-tugas biomedis, penggunaan filter analog pada frekuensi yang sangat rendah sulit, berbeda dengan filter digital.

Kekurangan Filter Digital:

- Keterbatasan resolusi. Dalam proses perhitungan pada filter digital dengan resolusi terbatas, terjadi akumulasi noise kuantisasi dan noise pembulatan. Salah satu tugas umum DSP adalah memilih resolusi filter yang sesuai.
- Keterbatasan kecepatan pemrosesan. Biasanya, filter analog mampu memproses data pada frekuensi yang sangat tinggi. Filter digital dibatasi oleh frekuensi sampling.
- Perangkat keras. Karena sinyal nyata dalam alam adalah kontinu, ADC dan DAC digunakan untuk memprosesnya. Kualitas sinyal digital (analog) yang dihasilkan juga tergantung pada karakteristik perangkat ini.

Sinyal dan filter didefinisikan oleh berbagai fungsi tergantung pada cara mereka direpresentasikan

| Sinyal | Filter |
|--------|--------|
|Sampel $x(n)$|Karakteristik Impuls $h(n)$|
|Bentuk Z $X(z)$|Fungsi Transfer $H(z)$|
|Kepadatan Spektral $X(e^{jΦ})$|Karakteristik Frekuensi $H(e^{jΦ})$|
|Spektrum Modulus $X(e^{jΦ})$|Respon Frekuensi $H(e^{jΦ})$|
|Spektrum Argumen $arg X(e^{jΦ})$|Respon Fase $arg H(e^{jΦ})$|i

Impuls Response (Respon Impuls) $h(n)$ adalah reaksi dari rangkaian terhadap masukan dalam bentuk fungsi Delta Dirac (fungsi δ). Ini mencerminkan pengaruh rangkaian pada sinyal. Dalam perangkat digital, respon impuls dapat menjadi berakhir atau tak terbatas. Oleh karena itu, ada dua kelas filter digital - dengan respon impuls berakhir dan tak berakhiral:

Filter digital dijelaskan oleh persamaan diferensial:

$$y(n) = \frac{1}{a_0} . (\sum\limits_{k=0}^{N} b_k x(n-k) - \sum\limits_{k=1}^{M} a_k y(n-k))$$

Dalam formula ini, $y(n)$ adalah respons keluaran atau sampel pada keluaran dari filter digital. $x(n)$ adalah sinyal masukan. $a_k$ dan $b_k$ adalah faktor pengali (koefisien) dari elemen-elemen digital baik pada masukan maupun keluaran. Faktor-faktor ini juga disebut sebagai koefisien pembilang dan penyebut dari karakteristik transfer filter.

Karakteristik Transfer filter memiliki bentuk berikut:

$$H(z) = \frac{B(z)}{A(z)} = \frac{b_0 + b_1 z^{-1} + ... + b_N z^{-N}}{1 + a_1 z^{-1} + ... + a_M z^{-M}}$$

Di mana N dan M adalah jumlah garis tunda untuk sinyal masukan dan keluaran. Secara umum, ini adalah formula untuk filter IIR. Jika penyebutnya sama dengan satu, maka rumus sesuai dengan ekspresi untuk filter FIR (filter digital tanpa umpan balik).

Dalam implementasi filter digital, ada dua metode yang berbeda: perangkat keras dan perangkat lunak. Filter perangkat keras diimplementasikan pada FPGA atau prosesor sinyal khusus. Filter perangkat lunak diimplementasikan menggunakan program yang dijalankan oleh prosesor atau mikrokontroler.

Filter bisa bersifat rekursif, di mana sampel keluaran tergantung pada sampel keluaran sebelumnya dan pada sampel masukan, yang berarti ada umpan balik dalam skema, atau non-rekursif, di mana sampel keluaran hanya tergantung pada sampel masukan.

## Filter IIR dan FIR

Filter FIR (finite impulse response) adalah filter digital dengan respon impuls yang terbatas. Karakteristik impuls filter ini terbatas dalam waktu, yang berarti memiliki jumlah koefisien yang terhitung. Mulai dari suatu titik waktu tertentu, karakteristik ini menjadi nol. Jika sinyal impuls satu satuan diberikan ke input filter FIR, output filter akan memiliki jumlah sampel yang terbatas. Biasanya, respons frekuensi filter dengan karakteristik impuls yang terbatas adalah linear.

Secara umum, filter FIR diimplementasikan tanpa umpan balik, artinya mereka non-rekursif. Namun, dengan menggunakan transformasi matematika, filter dapat diubah menjadi bentuk rekursif.

Persamaan Diferensial Filter FIR:

$$y(n) = \sum\limits_{k-0}^{N-1} h(k) x(n-k)$$

dimana

- $y(n)$ - sinyal diskrit output (jumlah dari puls input yang ditimbang),
- $x(n)$ - sinyal diskrit input (urutan sampel)
- $
h(n$) - koefisien karakteristik impuls filt
- N - panjang (orde) filter.

Karakteristik Transfer Filter FIR:

$$H(z) = \sum\limits_{k-0}^{N-1} h(k) z^{-k}$$

Peringatkan bahwa operasi $z^{-k}$ - merupakan penundaan urutan sebanyak k-sampel.

Filter IIR (infinite impulse response) adalah filter digital dengan karakteristik impuls yang tak terbatas dalam waktu, artinya memiliki jumlah koefisien yang sangat besar atau tak terbatas. Filter IIR juga disebut sebagai filter rekursif karena dalam implementasinya, umpan balik digunakan (sinyal dari output filter melalui elemen penundaan masuk kembali ke filter dan mempengaruhi perubahan di dalamnya). Fungsi transfer filter IIR memiliki bentuk pecahan-rasional. Filter IIR yang dikenal adalah filter Chebyshev, Butterworth, Kalman, Bessel, dan sebagainya.

Persamaan diferensial filter IIR:

$$y(n) = \sum\limits_{k=0}^{N} b_k x(n-k) - \sum\limits_{k=1}^{M} a_k y(n-k)$$

dan

$$y(n) = b_0 x(n) + b_1 x(n-1) + b_M x(n-M) + ... - a_1 y(n-1) - a_2 y(n-2) - ... - - a_N y(n-N)$$

Seperti yang terlihat, sinyal output mempengaruhi dirinya sendiri melalui umpan balik.

Karakteristik Transfer Filter IIR:

$$H(z) = \frac{B(z)}{A(z)} = \frac{b_0 + b_1 z^{-1} + ... + b_N z^{-N}}{1 + a_1 z^{-1} + ... + a_M z^{-M}}$$

Berbeda dengan filter FIR, filter IIR tidak selalu stabil. Untuk menjaga stabilitas filter digital IIR, semua akar fungsi transfer harus memiliki magnitudo kurang dari satu (artinya, berada di dalam lingkaran satuan di bidang z.

Orde filter - derajat tertinggi dalam ekspresi fungsi transfer $H(z$).

Fungsi transfer dapat diimplementasikan menggunakan berbagai skema struktural. Skema struktural yang paling umum adalah skema struktural serial dan paralel. Skema struktural serial umumnya digunakan ketika fungsi transfer dari filter rekursif dapat dipecah menjadi produk fungsi transfer orde pertama dan kedua. Skema struktural paralel, di sisi lain, digunakan ketika fungsi transfer dari filter rekursif dapat dipecah menjadi jumlah fungsi transfer orde pertama dan kedua.

## Skema Kanonik dan Skema Langsung

Struktur skema filter digital dapat direpresentasikan dalam bentuk langsung dan bentuk kanonikal. Filter digital dapat diimplementasikan menggunakan tiga elemen digital: pengalinya, penjumlah, dan blok penundaan. Jelas bahwa pengalinya diperlukan untuk mengalikan sampel dengan koefisien $a_k$ dan $b_k$, sementara penjumlah menggabungkan hasil yang diperoleh.

Skema struktural dari filter FIR non-rekursif sesuai dengan persamaan:

Jumlah elemen penunda selalu satu kurang dari jumlah koefisien atau orde filter. Karakteristik impuls filter FIR sesuai dengan kumpulan koefisien dari filter tersebut.

Skema struktural dari filter IIR terlihat lebih kompleks karena adanya umpan balik.

|

Jumlah elemen penunda pada filter IIR adalah $N + M$.

Skema struktural dalam bentuk kanonik mengurangi jumlah elemen penunda dengan menggunakan elemen penunda yang bersamaan. Penjumlah masukan menggunakan umpan balik. Penjumlah keluaran menggabungkan hasil. Transformasi dari satu bentuk ke bentuk lainnya cukup mudah dan sederhana.

## Pemilihan Filter

Pemilihan jenis filter sering menjadi pertimbangan penting bagi pengembang perangkat digital. Pilihan antara filter FIR dan IIR dilakukan berdasarkan kriteria berikut:

- Karakteristik fase dari filter IIR bersifat non-linear. Filter FIR dapat memiliki karakteristik fase yang secara ketat linear. Ini berarti filter tersebut tidak menyebabkan distorsi pada bentuk sinyal.
- Filter FIR stabil karena diimplementasikan dalam bentuk non-rekursif. Filter IIR dapat menjadi tidak stabil.
- Dalam kasus nyata, pertanyaan tentang pemilihan resolusi data (pembulatan, kuantisasi) menjadi kurang signifikan dalam perangkat tanpa umpan balik, yaitu dalam filter FIR.
- Implementasi respons frekuensi dengan bentuk yang kompleks atau maksimal persegi membutuhkan jumlah koefisien filter FIR yang signifikan. Filter IIR lebih efisien dalam menangani tugas ini.
- Filter FIR umumnya tidak memiliki filter analog yang setara.
- Filter FIR memungkinkan untuk dengan mudah mendapatkan karakteristik yang diinginkan (level penurunan, ketidakmerataan di jalur lewat, frekuensi cut-off, dll.)
- Filter IIR jauh lebih efisien dalam hal jumlah operasi perkalian, penambahan, dan jumlah elemen penundaannya.



### Contoh

Misalkan ada filter yang dijelaskan oleh fungsi transfer berikut:

$$H(z) = \frac{b_0 + b_1 z^{-1} + b_2 z^{-2}}{1 + a_1 z^{-1} + a_2 z^{-2}}$$

dimana

- $N = 2$, $M = 2$ 
- $b_i = {0.1, -0.6, 0.5}$
- $a_i = {1, 0.7, -0.2}$

Jumlah persamaan diferensial untuk filter tertentu sama dengan jumlah penjumlah pada skema. Dengan mengetahui persamaan diferensial, kita dapat menemukan karakteristik impuls filter: sinyal impuls satu unit diterapkan ke input (kondisi awalnya nol).

Kita akan menemukan karakteristik impuls menggunakan Python. Dengan menggunakan fungsi lfilter(b, a, x), proses melewati sinyal x melalui filter digital dipodelkan, dengan koefisien karakteristik transfer a dan b.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import lfilter, freqz, butter, firls, remez, firwin, firwin2, group_delay
from scipy.fftpack import fft, fftshift

FONT_SMALL = 12
FONT_MEDIUM = 14

plt.rc('axes', titlesize=FONT_MEDIUM) 
plt.rc('axes', labelsize=FONT_SMALL)
plt.rc('xtick', labelsize=FONT_SMALL)
plt.rc('ytick', labelsize=FONT_SMALL) 
plt.rc('legend', fontsize=FONT_MEDIUM) 
plt.rc('figure', titlesize=FONT_MEDIUM)

%matplotlib inline
```

```python
N = 128
# Delta-function
x = np.zeros(N)
x[0] = 1

# Input signal:
# h(z) = [b0 + b1*z^(-1) + b2*z^(-2)] / a0 + a1*z^(-1) + a2*z^(-2)
b = [0.1, -0.6, 0.5]
a = [1, 0.7, -0.2]

y1 = lfilter(b, a, x)

plt.figure(figsize=(16, 4), dpi=80)
plt.title('Impulse responce')
plt.stem(y1, basefmt='C0')
plt.xlim([-0.5, N-0.5])
plt.grid(True)
```

Jika filter stabil, maka nilai-nilai y(n) akan berkurang seiring waktu.

Seperti yang terlihat pada grafik sebelumnya, filter IIR yang dihitung adalah stabil.

### Filter Butterworth

Contoh implementasi filter Butterworth orde 3. Kami akan menunjukkan bagaimana sinyal yang bising melewati filter tersebut:

```python
# Create input signal
t = 2 * np.pi * np.linspace(-1, 1, 500)
x = np.sin(0.25*t*t)+ 0.95*np.sin(2.0*t)

# Add some white noise
np.random.seed(1)
xn = x + np.random.randn(len(t))

# 3-order lowpass butterworth filter
b, a = butter(3, 0.2)
z = lfilter(b, a, xn)

wn = [0.01, 0.05, 0.1, 0.2]

# Calculate IIR filter
zz = np.zeros((t.size, 4))
for i in range(4):
    b, a = butter(3, wn[i])
    zz[:, i] = lfilter(b, a, xn)

# Plot results
plt.figure(figsize=(16, 8), dpi=80)
for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.plot(t, xn, 'C0--', linewidth=1.5)
    plt.plot(t, zz[:,i], 'C1', linewidth=2.5)
    plt.xlim([-2 * np.pi, 2 * np.pi])
    plt.grid(True)
    plt.legend(('Signal', 'Filtered, wn = {}'.format(wn[i])), loc='lower right')
```

### Filter FIR Rekursif

Seperti yang disebutkan sebelumnya, filter FIR biasanya dibangun dalam bentuk non-rekursif, namun ada cara untuk mengubah filter menjadi bentuk rekursif.

Mari kita lihat contoh berikut. Mari kita tulis karakteristik transfer filter:

$$H(z) = 1 + z^{-1} + z^{-2} + z^{-3}$$

Kita akan mengalikan fungsi transfer dengan

$$\frac{1-z^{-1}}{1-z^{-1}}$$

Kami akan melewati langkah-langkah matematis (jika Anda ingin, Anda bisa melakukannya sendiri karena sifat trivial dari masalah yang dipertimbangkan) dan menulis hasilnya:

$$H(z) = \frac{1-z^{-3}}{1-z^{-1}}$$

Skema struktural dari filter tersebut ditunjukkan pada gambar berikut. Ini terdiri dari dua rantai: elemen tanpa umpan balik - yang diferensiasi, dan elemen dengan umpan balik - yang mengintegrasi (penjumlah dengan umpan balik).





Mari kita buat ilustrasi bahwa sinyal yang melewati filter FIR rekursif memiliki karakteristik yang serupa.

```python
N = 40
# Delta-function
x = np.zeros(N)
x[0] = 1

# Filter order - M
M = 16
# Input signal
# h(z) = 1 + z^(-1) + ... + z^(-9)
b = np.ones(M-1)
a = [1]

y1 = lfilter(b, a, x)

# Change to recursive form
b = np.zeros(M)
b[0] = 1
b[M-1] = -1

a = [1, -1]
y2 = lfilter(b, a, x)

# Check the difference
np.all(y1 == y2)
```

```python
plt.figure(figsize=(16, 3), dpi=80)
plt.stem(y2, basefmt='C0')
plt.xlim([-0.5, N-0.5])
plt.xlabel('Samples')
plt.suptitle('Filtered signal')
plt.grid(True)
```

Seperti yang terlihat dari grafik, karakteristik impuls filter rekursif adalah vektor satu. Selain itu, panjang vektor sama dengan panjang filter.

### Filter FIR Homogen

Filter FIR yang disebut homogen adalah filter FIR yang tidak memerlukan operasi perkalian. Oleh karena itu, filter tersebut bekerja cukup cepat dan dapat diimplementasikan dengan baik di FPGA.

Kekurangan dari filter semacam itu adalah tingkat penekanannya yang tinggi pada sisi gelombang. Sisi gelombang dalam respons frekuensi ditekan dengan mengkaskadkan filter, di mana posisi nol tidak berubah, tetapi hanya perbandingan antara level dari sisi utama dan sisi gelombang yang berubah, sehingga level relatif sisi gelombang menurun ketika respons frekuensi dikalikan.

Kekurangan kedua dari filter homogen adalah bentuk tidak persegi respons frekuensinya. Untuk membuatnya lebih persegi, filter FIR korektif digunakan, sehingga bentuk dari puncak utamanya menjadi lebih persegi, tetapi tingkat sisi gelombang meningkat.

Pada bagian yang terkait dengan decimation dan interpolation dari sinyal, kita akan menjelajahi lebih lanjut tentang filter ini.

Anda dapat menggunakan fungsi freqz untuk menemukan karakteristik frekuensi filter.

```python
N = 512
f = np.linspace(0, 1, N, endpoint=True)
b = [1, 1, 1]

plt.figure(figsize=(16, 4))
plt.subplot(1, 2, 1)
for i in range(4):
    _, h = freqz(b, N)
    h = np.abs(h)
    plt.plot(f, h/np.max(h))
    plt.xlim([0, 1])
    plt.grid(True)
    b.append(1)
plt.xlabel('Normalized Freq')
plt.title('CIC freq responce')
plt.tight_layout()
```

### Filter FIR dengan Phase Linear

Fitur paling penting dari filter FIR adalah kemampuannya untuk memberikan karakteristik fase linear yang tepat.

Mengapa ini penting? Saat sinyal melewati filter, berbagai transformasi terjadi. Secara khusus, amplitudo dan fase sinyal berubah tergantung pada karakteristik frekuensi filter (amplitudo dan fase, FCX). Untuk sinyal multiflektrum, penting bagi fase sinyal tetap tidak terdistorsi saat melewati simpul pemrosesan digital. Sementara membuat ACX hampir konstan di jalur lintasan bukanlah masalah, masalah muncul dengan FCX.

Untuk mengevaluasi distorsi fase, berguna untuk memperkenalkan konsep delay fase dan grup.

Delay fase adalah jumlah delay untuk setiap komponen frekuensi sinyal. Ini didefinisikan sebagai sudut pergeseran fase dibagi dengan frekuensi.

$$T_{\phi} = \frac{-\theta(\omega)}{\omega}$$

Grup Delay adalah rata-rata penundaan waktu dari seluruh sinyal multiflektrum. Didefinisikan sebagai turunan fase terhadap frekuensi.

$$T_{g} = \frac{-d\theta(\omega)}{\omega}$$

Dari rumus untuk Grup Delay, kondisi linearitas FCX filter menjadi jelas. Aturannya:

Jika FCX adalah linear, maka Grup Delay setelah diferensiasi sama dengan konstan.

Dengan kata lain, Grup Delay konstan untuk semua komponen frekuensi. Dengan demikian, linearitas karakteristik fase adalah salah satu fitur paling penting dari filter FIR.

### Simetri Karakteristik Impuls

Untuk memastikan linearitas FCX, kondisi simetri dari karakteristik impuls (atau koefisien) filter harus dipenuhi. Dengan kata lain, filter FIR dengan FCX linear adalah simetris. Ada 4 jenis filter yang berbeda dalam hal kelipatan N dan jenis simetri:

- simetri positif, kelipatan genap,
- simetri negatif, kelipatan genap,
- simetri positif, kelipatan ganjil,
- simetri negatif, kelipatan ganjil.

Sebagai contoh, untuk filter dengan simetri negatif, kita dapat mencapai pergeseran fasa sebesar $\frac{\pi}{2}$. Filter semacam itu digunakan untuk merancang diferensiator dan transformasi Hilbert.

Envelope dari karakteristik impuls filter FIR dibangun berdasarkan hukum $\frac{sin(x)}{x}$ terlepas dari jenis filter (rendah, tinggi, diferensiator, filter pita, atau filter penolak). Untuk menyelesaikan masalah praktis, tidak perlu memikirkan jenis filter yang dipilih.h.

## Desain Filter FIR

Dalam kebanyakan kasus, "perhitungan filter FIR" mengacu pada pencarian koefisien filter berdasarkan nilai-nilai karakteristik frekuensi.

Desain filter FIR digital melalui beberapa tahapan pengembangan:

- Spesifikasi filter. Jenis filter ditentukan (LPF, HPF, bandpass, rejector), jumlah koefisien N, karakteristik frekuensi yang diinginkan, dengan toleransi non-linearitas di dalam band-stop dan band-pass, dll.
- Perhitungan koefisien. Koefisien filter yang memenuhi spesifikasi dari langkah sebelumnya dihitung menggunakan metode dan alat yang tersedia.
- Analisis dampak akurasi akhir. Pada tahap ini, pengaruh efek kuantisasi terhadap koefisien filter, data sementara, dan output dievaluasi.
- Implementasi. Pada tahap ini, pengembangan filter dilakukan dengan menggunakan perangkat keras atau perangkat lunak: dalam bahasa pemrograman yang tersedia atau berbasis FPGA atau prosesor sinyal khusus.

Tahapan pengembangan mungkin sedikit berbeda, tetapi esensi dari desain filter FIR digital selalu tetap sama.

### Spesifikasi filter

Pada tahap ini, insinyur mencari solusi kompromi untuk mewujudkan filter FIR yang diinginkan dengan parameter yang dibutuhkan. Ada beberapa parameter yang harus dipertimbangkan, dan seringkali satu parameter harus dikorbankan untuk mencapai nilai yang diinginkan untuk parameter lainnya.

- $A_{pass}$ - ketidakseragaman di dalam jalur lulus,
- $A_{stop}$ - level penurunan di dalam jalur penolakan,
- $F_{pass}$ - frekuensi batas jalur lulus,
- $F_{stop}$ - frekuensi batas jalur penolakan,
- $N$ - urutan filter (jumlah koefisien filter).

Dalam praktiknya, parameter $A_{pass}$ dan $A_{stop}$ didefinisikan dalam desibel (dB), sedangkan jarak antara $F_{pass}$ dan $F_{stop}$ menunjukkan lebar jalur transisi filter. Secara logis, nilai $A_{pass}$ harus seminim mungkin, $A_{stop}$ semaksimal mungkin, dan rasio $\frac{F_{pass}}{F_{stop}}$ idealnya mendekati satu (penampang frekuensi yang ideal berbentuk persegi).

Jumlah koefisien tidak disertakan dalam spesifikasi filter tanpa alasan. Seperti yang akan ditunjukkan nanti, parameter-parameter karakteristik frekuensi filter bergantung pada urutan filter N dan resolusi koefisien.

### Perhitungan Koefisien Filter

Ada banyak metode untuk menghitung koefisien filter - metode pemutaran jendela, metode sampling frekuensi, berbagai metode optimal (seperti Chebyshev) dengan menggunakan algoritma Remez, dan sebagainya. Setiap metode memiliki karakteristik uniknya sendiri dan memberikan hasil yang berbeda-beda. Untuk metode pemutaran jendela, efek negatif yang muncul adalah fenomena Gibbs, yang menghasilkan ketidakerataan dan lonjakan dalam karakteristik frekuensi filter di antara titik-titik yang dihitung. Fenomena ini dapat diatasi dengan memberikan toleransi pada ketidakerataan di dalam jalur transmisi dan pita penolakan.

Metode utama untuk menghitung koefisien filter pada banyak kasus adalah algoritma Parks-McClellan, juga dikenal sebagai algoritma Remez yang dimodifikasi. Algoritma ini adalah metode iteratif tidak langsung untuk menemukan nilai-nilai optimal dengan karakteristik Chebyshev filter. Keunikan dari metode ini adalah minimisasi kesalahan di jalur transmisi dan pita penolakan melalui pendekatan Chebyshev terhadap karakteristik impuls.

Secara logis, semakin banyak jumlah koefisien, semakin sedikit ketidakerataan di karakteristik frekuensi dan semakin dekat dengan bentuk persegi di dalam pita transmisi. Pilihan metode sangat mempengaruhi hasil akhir, tetapi semuanya bertujuan pada hal yang sama - meminimalkan lonjakan di dalam pita transmisi dan meningkatkan "kepersegiannya" karakteristik frekuensi (dengan tetap mempertahankan kondisi linieritas Fase).

### Analisis Konsekuensi dari Ketepatan Jumlah Bit yang Terbatas

Kedalaman bit koefisien adalah faktor utama yang memengaruhi bentuk karakteristik frekuensi. Sebagai contoh, pada rangkaian terpadu logika yang dapat diprogram (PLD) modern, kedalaman bit koefisien dapat dipilih secara bebas, tetapi nilai yang sebenarnya berada dalam kisaran 16 hingga 27 bit karena sifat-sifat blok komputasi (DSP48) yang diimplementasikan.

Untuk filter-order yang tinggi, sering kali diperlukan rentang dinamis yang besar dalam grid bit, tetapi jika itu tidak dapat dicapai, akhirnya akan muncul kesalahan kuantisasi. Karena keterbatasan resolusi bit koefisien, karakteristik frekuensi dimodifikasi, dan dalam beberapa kasus, distorsi begitu kuat sehingga harus mengorbankan parameter dari spesifikasi frekuensi untuk mencapai hasil yang dapat diterima. Dengan kata lain, resolusi representasi koefisien secara langsung mempengaruhi penurunan maksimum yang mungkin dicapai $A_{stop}$. Oleh karena itu, ketika menggunakan grid bit yang terlalu terbatas, terkadang tidak mungkin mencapai penekanan yang diinginkan bahkan dengan order filter yang sangat besar!

### Perhitungan Filter Konvolusi Tak Hingga dengan Metode Kuadrat Terkecil

Dengan menggunakan metode kuadrat terkecil, fungsi ini menghitung koefisien filter yang memiliki pendekatan terbaik terhadap karakteristik frekuensi yang dibutuhkan. Secara matematis, dalam mencari koefisien, integral dari rata-rata kuadrat error yang ditimbang dalam rentang bandwidth akan diminimalkan.

Metode `firls` mengembalikan koefisien filter FIR. Argumen input dari fungsi:

- numtaps - orde filter,
- bands - urutan monoton tidak menurun dari frekuensi. Semua elemen harus non-negatif dan kurang dari atau sama dengan frekuensi Nyquist, yang ditentukan oleh parameter nyq.
- desired - urutan amplitudo, berisi penguatan dan penurunan yang diinginkan pada nilai frekuensi yang ditentukan dalam bands.
- weight - "bobot" relatif dari bandwidth dan penekanan. Menentukan kompromi kualitas di bandwidth-bandwidth tersebut.
- nyq atau fs - Frekuensi Nyquist. Jika tidak ditentukan, defaultnya adalah satu.

### Perhitungan Filter Konvolusi Tak Hingga dengan Algoritma Remez:

Metode `remez` mengembalikan koefisien filter FIR. Argumen input dananya identik dengan metode sebelumnya:

- numtaps - orde filter,
- bands - urutan monotonik tidak menurun dari frekuensi. Semua elemen harus non-negatif dan kurang dari atau sama dengan frekuensi Nyquist, yang ditentukan oleh parameter Hz.
- desired - urutan amplitudo, berisi penguatan dan penurunan yang diinginkan pada nilai frekuensi yang ditentukan dalam bands.
- weight - "bobot" relatif dari bandwidth dan penekanan. Menentukan kompromi kualitas di bandwidth-bandwidth tersebut.
- Hz atau fs - Frekuensi Nyquist. Jika tidak ditentukan, defaultnya adalah satu.

### Perhitungan Filter Konvolusi Tak Hingga dengan Metode Jendela:
Metode firwin dan firwin2 ` mengembalikan koefisien filter FIR.

#### Metode firwin:

- numtaps - urutan filter,
- cutoff - frekuensi cutoff relatif terhadap laju sampel fs.
- width - jika nilai ditentukan, menunjukkan lebar transisi dari jalur lintasan ke jalur penekanan.
- window - fungsi jendela. Digunakan bersamaan dengan metode get_window atau langsung menetapkan nama jendela.
- fs - laju sampel atau frekuensi Nyquist. Jika nilai tidak ditentukan, secara default sama dengan s#### atu.

Metode firwin2:

- numtaps - urutan filter,
- freq - urutan frekuensi monotonik tak menurun. Semua elemen harus non-negatif dan kurang dari atau sama dengan frekuensi Nyquist yang ditentukan oleh parameter fs.
- gain - urutan amplitudo yang berisi penguatan dan redaman yang diinginkan pada nilai frekuensi yang ditentukan dari freq.
- window - fungsi jendela. Digunakan bersamaan dengan metode get_window atau langsung menetapkan nama jendela.
- fs - laju sampel atau frekuensi Nyquist. Jika nilai tidak ditentukan, secara default sama dengan satu.

### Fungsi jendela

### Gelombang grup

Ini adalah fungsi bantuan untuk menghitung penundaan grup filter digital. Parameter masukan - larik koefisien a, b. Metode mengembalikan kumpulan frekuensi dan larik penundaan grup yang sesuai untuk frekuensi tersebut.

### Contoh

Contoh pertama - menggunakan fungsi firwin2. Mari tentukan larik frekuensi yang dinormalisasi dari 0 hingga 1. Tentukan penguatan pada frekuensi yang dipilih, serta urutan filter. Mari kita buat karakteristik impuls.

```python
N = 100
# List of freqs
lst_freqs = np.linspace(0, 1, N)

# List of gains
lst_gain = np.zeros(N)
lst_gain[0:20] = 1

# FIR filter taps
NFIR = 64
taps = firwin2(NFIR, lst_freqs, lst_gain)

plt.figure(figsize=(16, 4), dpi=100)
plt.title('Impulse responce')
plt.stem(taps, basefmt='C0')
plt.xlim([0, NFIR-1])
plt.xlabel('Samples')
plt.grid(True)
```

Mari kita buat karakteristik amplitudo-frekuensi (AF) dari filter pita dan filter turun rendah menggunakan tiga metode: firls(), remez(), dan firwin2().

```python
# Input parameters
fs = 10
N = 117
desired = (0, 0, 1, 1, 0, 0)
bands = (0, 1, 2, 4, 4.5, 5)

# FIR filters
fir_firls = firls(N, bands, desired, fs=fs)
fir_remez = remez(N, bands, desired[::2], fs=fs)
fir_firwin2 = firwin2(N, bands, desired, fs=fs)

# PLot results and calculate FFTs
plt.figure(figsize=(16, 5))
plt.title('Frequency responce')
for fir in (fir_firls, fir_remez, fir_firwin2):
    freq, resp = freqz(fir)
    resp = np.abs(resp)
    resp /= np.max(resp) + 10**(-15)
    plt.plot(freq, 20*np.log10(resp))
plt.xlim([0, np.pi])
plt.ylim([-180, 5])
plt.legend(['firls', 'remez', 'firwin2'], loc ='upper left')
plt.grid(True)
```

### Kaiser FIR Filter

Mari kita lihat bagaimana sinyal harmonik yang bising melewati filter FIR. Kita akan menambahkan beberapa sinyal harmonik frekuensi tinggi dengan amplitudo yang berbeda ke sinyal ini.

Seperti yang terlihat, sinyal bising tidak dapat dianalisis dengan baik dalam domain waktu. Namun, dengan melewatkannya melalui filter FIR dengan frekuensi cut-off yang dipilih sebelumnya, kita dapat memisahkan sinyal yang menarik dari gangguan.

Perlu dicatat bahwa semua sinyal dan noise yang masuk ke dalam jalur lintasan filter akan tetap ada dan akan berkontribusi pada karakteristik kualitatif sinyal hasil.

```python
N = 1024
# Create input signal
t = np.linspace(0, 1, N, endpoint=True)
x = 3*np.cos(2*np.pi*20*t) + 210*np.sin(2*np.pi*117*t) + 100*np.sin(2*np.pi*184*t) + 380*np.sin(2*np.pi*281*t)

# Add some white noise
np.random.seed(1)
xn = x + 5*np.random.randn(N)

# Kaiser FIR filter
taps = 160
h = firwin(taps, 0.1, window=('kaiser', 9))
y = lfilter(h, 1, xn)

# List of input signals
lst_sig = [xn, h, y]

sig_titles = ['Noisy signal', 'FIR impulse responce', 'Filtered signal']
fft_titles = ['Input spectrum', 'FIR filter', 'Result']


plt.figure(figsize=(16, 10))
for i in range(3):
    # Calculate FFTs
    clc_fft = np.abs(fft(lst_sig[i], N)) 
    clc_fft = 20*np.log10(10e-11+clc_fft/np.max(clc_fft))
    
    # Plot signals
    plt.subplot(3, 2, 2*i+1)
    plt.plot(lst_sig[i], color='C'+str(i))
    plt.title(sig_titles[i])
    if (i == 2):
        plt.ylim([-5, 5])
        plt.xlim([taps, lst_sig[i].size//2-1])
    else:
        plt.xlim([0, lst_sig[i].size-1])
    plt.grid(True)
    
    plt.subplot(3, 2, 2*(i+1))
    plt.plot(clc_fft, color='C'+str(i))
    plt.title(fft_titles[i])
    plt.xlim([0, N//2-1])
    plt.grid(True)
plt.tight_layout()
```

Pada gambar di atas, terlihat bahwa sinyal keluaran yang difilter dengan sedikit distorsi menyerupai sinyal harmonik yang diinginkan. Namun, disayangkan bahwa mencapai bentuk gelombang harmonik yang ideal melalui filtrasi FIR atau IIR tidak mungkin, karena tujuan utama dari filter adalah untuk memotong komponen spektral yang tidak diinginkan. Untuk meningkatkan kualitas sinyal yang difilter, metode pemrosesan sinyal digital lainnya (spektral dan korelasi) diterapkan.

Ini mengakhiri bagian yang berkaitan dengan filtrasi digital. Pada bagian-bagian berikutnya, akan dijelaskan secara rinci tentang fungsi jendela, tugas pemrosesan sinyal multi-kecepatan - dekimasikan dan interpolasi, perhitungan filter-korektor, dan sebagainya.
